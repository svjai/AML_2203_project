```python
headers = {
    'sec-ch-ua': '^\\^Not',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
    'Referer': 'https://www.ebooks.com/en-ca/subjects/computers/',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
    'Content-Type': 'application/json',
}

title=[]
id_=[]
price=[]
author=[]
publisher=[]
pub_year=[]
sub=[]
edition=[]
description=[]
availablity=[]
sale_date=[]
short_pub_date=[]
num_of_author=[]
c_image_width=[]
c_image_height=[]



```

```python
#final run
#removed all the cells (it was very big and i used many things which will be confusing)
#trauncated all the experiments and added the final version (just for your understanind)
```

```python
for page in range(1,4000):
    
    params = (
        ('CountryCode', 'CA'),
        ('subjectId', '13'),
        ('pageNumber',str(page))
    )

    response = requests.get('https://www.ebooks.com/api/search/subject/', headers=headers, params=params)
    result=response.json()
    res=result['books']
    for book in res:
        title.append(book['title'])
        id_.append(book['id'])
        price.append(book['price_number'])
        try:
            author.append(book['authors'][0]['name'])
        except IndexError:
            author.append(None)
        publisher.append(book['publisher'])
        pub_year.append(book['publication_year'])
        sub.append(book['subtitle'])
        edition.append(book['edition'])
        description.append(book['description'])
        availablity.append(book['is_available'])
        sale_date.append(book['on_sale_date'])
        short_pub_date.append(book['short_publication_date'])
        num_of_author.append(book['num_authors'])
        c_image_width.append(book['image_width'])
        c_image_height.append(book['image_height'])

    
```

```python
df3=pd.DataFrame({'Title':title,'id':id_,'price':price,'author':author,'publisher':publisher,'pub_year':pub_year,'s_tile':sub,'edition_num':edition,'description':description,'availabe':availablity,'sale_date':sale_date,'short_pub':short_pub_date,'num_of_author':num_of_author,'width':c_image_width,'height':c_image_height})
```

```python
df3.to_csv('api_metod_40k_15_fields_vahid.csv',index=False)
```
