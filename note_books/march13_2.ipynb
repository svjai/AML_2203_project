```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import MiniBatchKMeans, KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('api_metod_40k_15_fields.csv')
```

```python
df.dropna(subset=['Title'], inplace=True)
```

```python
# Preprocessing the titles: lowercase and remove punctuation
df['processed_title'] = df['Title'].str.lower().str.replace('[^\w\s]', '', regex=True)
```

```python
# Outlier detection based on title length
df['title_length'] = df['processed_title'].apply(len)
lower_threshold = df['title_length'].quantile(0.05)
upper_threshold = df['title_length'].quantile(0.95)
df_filtered = df[(df['title_length'] >= lower_threshold) & (df['title_length'] <= upper_threshold)]
```

```python
# Vectorization using CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df_filtered['processed_title'])
X_norm = normalize(X)  # Normalizing the data
```

```python
# Determine the optimal number of clusters using Elbow Method and Silhouette Score
range_n_clusters = range(2, 21)
sse = []
silhouette_scores = []
```

```python
for n_clusters in range_n_clusters:
    minibatch_kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=42, batch_size=100)
    minibatch_kmeans_labels = minibatch_kmeans.fit_predict(X_norm)
    sse.append(minibatch_kmeans.inertia_)
    
    # Compute the silhouette score
    if 1 < n_clusters < X_norm.shape[0] - 1:  # Silhouette score requires more than one cluster
        silhouette_avg = silhouette_score(X_norm, minibatch_kmeans_labels)
        silhouette_scores.append(silhouette_avg)
        print(f"For n_clusters = {n_clusters}, the silhouette score is : {silhouette_avg}")

# Plot SSE for Elbow Method
plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, sse, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()

# Plot Silhouette Scores
plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Silhouette Score For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
optimal_n_clusters = 15
```

```python
# Run MiniBatch K-Means with the optimal number of clusters
optimal_minibatch_kmeans = MiniBatchKMeans(n_clusters=optimal_n_clusters, random_state=42, batch_size=100)
optimal_minibatch_kmeans_labels = optimal_minibatch_kmeans.fit_predict(X_norm)
df_filtered['Cluster_MiniBatchKMeans'] = optimal_minibatch_kmeans_labels
```

```python
# Print out the first 10 titles in each cluster for MiniBatch K-Means
for i in range(optimal_n_clusters):
    cluster_indices = df_filtered['Cluster_MiniBatchKMeans'] == i
    titles_in_cluster = df_filtered.loc[cluster_indices, 'Title'].head(10).tolist()
    print(f"\nMiniBatch K-Means Cluster {i}:")
    print('\n'.join(titles_in_cluster))
```

```python
# Apply K-Means with the optimal number of clusters found
optimal_kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42)
optimal_kmeans_labels = optimal_kmeans.fit_predict(X_norm)
df_filtered['Cluster_KMeans'] = optimal_kmeans_labels
```

```python
# Print out the first 10 titles in each cluster for K-Means
for i in range(optimal_n_clusters):
    cluster_indices = df_filtered['Cluster_KMeans'] == i
    titles_in_cluster = df_filtered.loc[cluster_indices, 'Title'].head(10).tolist()
    print(f"\nK-Means Cluster {i}:")
    print('\n'.join(titles_in_cluster))
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # eps and min_samples may need tuning
dbscan_labels = dbscan.fit_predict(X_norm)
df_filtered['Cluster_DBSCAN'] = dbscan_labels
```

```python
# Count the number of clusters formed by DBSCAN (excluding noise points labeled as -1)
n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print(f"DBSCAN formed {n_clusters_dbscan} clusters (excluding noise).")
```

```python
# Print out the first 10 titles in each cluster for DBSCAN
for i in range(n_clusters_dbscan):
    cluster_indices = df_filtered['Cluster_DBSCAN'] == i
    titles_in_cluster = df_filtered.loc[cluster_indices, 'Title'].head(10).tolist()
    print(f"\nDBSCAN Cluster {i}:")
    print('\n'.join(titles_in_cluster))
```
