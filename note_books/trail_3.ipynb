```python
import pandas as pd
from youtubesearchpython import VideosSearch

def search_youtube(query, max_results=10):
    videos_search = VideosSearch(query, limit=max_results)
    results = videos_search.result()['result']
    
    # List to store video details
    video_details = []
    
    for result in results:
        video_detail = {
            "title": result['title'],
            "video_id": result['id'],
            "channel_name": result['channel']['name'],
            "views": result['viewCount']['short'],
            "duration": result['duration'],
            "url": result['link']
        }
        video_details.append(video_detail)
    
    return video_details

# Example usage
search_queries = ["Python tutorial", "Python programming", "Python basics", "Python advanced","programming","code","python"]
max_results_per_query = 2000 # Adjust as needed

# Initialize an empty list to store all search results
all_search_results = []

for query in search_queries:
    search_results = search_youtube(query, max_results_per_query)
    all_search_results.extend(search_results)

# Create a pandas DataFrame
df = pd.DataFrame(all_search_results)

print("DataFrame:")
df.to_csv('you_tube_data.csv')
```
